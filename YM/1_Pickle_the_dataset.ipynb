```python
def pickle_chunks(path_to_file, chunksize = 10 ** 4, year = 'replace_me', compression = 'infer'):
    '''
    '''
    with pd.read_json(path_to_file, lines=True, compression='bz2', chunksize=chunksize) as df_reader:
        for index, chunk in enumerate(df_reader):
            pickle_file_name = year + '/' + year + '_chunk_' + str(index) + '.pkl'
            chunk.to_pickle(PICKLE_FOLDER + pickle_file_name, compression = compression)
    return print(pickle_file_name,'complete')
```

```python
def operation_on_pickled_chunk(df):
    return print(df.head())

def load_from_pickled_chunks(year = '2020'):
    directory = PICKLE_FOLDER + year + '/'
    for filename in os.listdir(directory):
        df = pd.read_pickle(directory + filename)
        operation_on_pickled_chunk(df)
        print(filename)
```

```python
def select_quotes_chunk(chunk, keywords):
    print(f'Processing chunk with {len(chunk)} rows')
    return chunk[chunk['quotation'].str.contains('|'.join(keywords))]

def select_from_pickled_chunks(keywords, year = '2020'):
    directory = PICKLE_FOLDER + year + '/'
    for filename in os.listdir(directory):
        df = pd.read_pickle(directory + filename)
        selected_df = select_quotes_chunk(df, keywords)
        print(selected_df.head(2))
```
