selecting only the first qid in qids column of sample data

```python
def first_qid(x):
    if x:
        x = x[0]
    else:
        x = '0'
    return x

def clean_qids(df):
    return df.apply(first_qid)
quotebank_2019_sample_clean =  quotebank_2019_sample.copy()
quotebank_2019_sample_clean['qids'] = clean_qids(quotebank_2019_sample['qids'])
```

```python
quotebank_2019_sample_clean.head()
```

```python
merged_df = quotebank_2019_sample_clean.merge(parquet_df, left_on = 'qids', right_on = 'id')
```
